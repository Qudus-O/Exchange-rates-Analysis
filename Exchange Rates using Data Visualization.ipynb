# Story Telling Data Visualization on Exchange Rates
The data that we'll be using shows the daily exchange rates in euros from 1999 to 2021. In the majority of European Union member states, the euro, denoted by the sign €, serves as the official currency.
One euro is worth more than one US dollar at this exchange rate, therefore if the euro to US dollar is worth 1.5, you will receive 1.5 US dollars for every 1.0 euro that you pay.
The data source is the European Central Bank, and Daria Chemkaeva assembled the set and made it accessible on [Kaggle](https://www.kaggle.com/lsind18/euro-exchange-daily-rates-19992020). We downloaded the dataset in January 2021; take note that it is regularly updated.

# Data Importation
The dataset will be read into a variable called `exchange_rates`.

```python
import pandas as pd
exchange_rates= pd.read_csv('euro-daily-hist_1999_2020.csv')
```

```python
exchange_rates.head() #first-five rows
```

```python
exchange_rates.tail()  #last-five rows
```

```python
exchange_rates.info()
```

```python
exchange_rates.rename(columns={'[US dollar ]': 'US_dollar',
                               'Period\\Unit:': 'Time'},
                      inplace=True)
exchange_rates['Time'] = pd.to_datetime(exchange_rates['Time'])
exchange_rates.sort_values('Time', inplace=True)
exchange_rates.reset_index(drop=True, inplace=True)

euro_to_dollar= exchange_rates[['Time','US_dollar']]
exchange_rates['US_dollar'].value_counts()
```

```python
euro_to_dollar = euro_to_dollar[euro_to_dollar['US_dollar'] != '-']
euro_to_dollar['US_dollar'] = euro_to_dollar['US_dollar'].astype(float)
euro_to_dollar.info()
```

# Rolling Mean

```python
import matplotlib.pyplot as plt
%matplotlib inline

euro_to_dollar['rolling_mean']= euro_to_dollar['US_dollar'].rolling(30).mean()
plt.plot(euro_to_dollar['Time'],euro_to_dollar['rolling_mean'])
plt.show()
```
